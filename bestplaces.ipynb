## Preprocessing ( Parsing , Geolocalizion adresses ..)

```python
colazione = colazione.split(sep = '\n')
```

```python
geolocator = Nominatim(user_agent="breakfast_rome")
```

```python
def parse_adresses(places:list , where = ' Rome, Italy'):
    dict_places = {}
    for p in places:
        
        res = geolocator.geocode( p + where)
        if res :
            dict_places[p]=res.point
        else :
            dict_places[p]=[]
    return dict_places
```

```python
# sanity check
for k in bar_di_roma.keys():
    if bar_di_roma[k] == []:
        print(k)
```

```python
bar_di_roma = parse_adresses(colazione)
```

```python
# check for empty
for k in bar_di_roma.keys():
    if bar_di_roma[k] == []:
        print(k)
```

```python
new = {}
for bar in just_address.split('\n'):
        
    res = geolocator.geocode( bar +' Rome, Italy')
    if res :
        new[bar]=res.point
    else :
        new[bar]=[]
```

```python
bar_di_roma.update(new)
```
