```python
import geopandas as gpd
import  folium 
import numpy as np
import matplotlib.pyplot as plt
```

import places from google

```python
where = './stars/Saved Places.json'
places = gpd.read_file(where)
```

```python
m = folium.Map(location=(52.518611, 13.408333),tiles = 'CartoDB positron', zoom_start=10)
for point in places.geometry:
    folium.Marker(location = (point.y , point.x)).add_to(m)
m
```

### manually assigning a color to every saved place 

```python
#first 60
to_append = []
for i, row in places.iloc[90:150,:].iterrows():
    location = (row.geometry.y,row.geometry.x)
    m = folium.Map(location= location, zoom_start=10)
    folium.Marker(location = location).add_to(m)
    
    
    m.save('save.html')
    to_append.append(input('What is this ?'))
```

```python
colors = np.concatenate([colors,np.array(to_append)])
np.save('colors',colors)
```
