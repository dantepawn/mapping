```python
import folium
from folium import plugins
```

```python
import numpy as np

data = (np.random.normal(size=(100, 3)) *
        np.array([[1, 1, 1]]) +
        np.array([[48, 5, 1]])).tolist()
```

`list of points of the form [lat, lng] or [lat, lng, weight]`

```python
data[:3]
```

```python
mapa = folium.Map([48., 5.], tiles='stamentoner', zoom_start=6)
mapa.add_child(plugins.HeatMap(data))
mapa
```

```python
mapa = folium.Map([48., 5.], tiles='stamentoner', zoom_start=6)
mapa.add_child(plugins.HeatMap(data , gradient = {0.4: 'blue', .65: 'lime',1 : 'yellow' }))
mapa
```
